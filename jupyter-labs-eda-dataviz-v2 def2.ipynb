### TASK  3: Visualize the relationship between success rate of each orbit type


Next, we want to visually check if there are any relationship between success rate and orbit type.


Let's create a `bar chart` for the sucess rate of each orbit


```python
# HINT use groupby method on Orbit column and get the mean of Class column
# Agrupamos por órbita y calculamos la tasa de éxito promedio
orbit_success = df.groupby("Orbit")["Class"].mean().reset_index()

# Gráfico de barras
sns.barplot(x="Orbit", y="Class", data=orbit_success)
plt.xlabel("Orbit Type", fontsize=14)
plt.ylabel("Success Rate", fontsize=14)
plt.title("Success Rate by Orbit Type", fontsize=16)
plt.xticks(rotation=45)
plt.show()
```

Analyze the ploted bar chart try to find which orbits have high sucess rate.
R: Las órbitas LEO e ISS tienen consistentemente altas tasas de éxito, lo que indica que los lanzamientos hacia estas órbitas suelen ser más controlados y técnicamente menos exigentes. En cambio, órbitas como GTO tienden a presentar más riesgos o menor tasa de éxito.

### TASK  4: Visualize the relationship between FlightNumber and Orbit type


For each orbit, we want to see if there is any relationship between FlightNumber and Orbit type.


```python
# Plot a scatter point chart with x axis to be FlightNumber and y axis to be the Orbit, and hue to be the class value
# TASK 4: Visualize the relationship between FlightNumber and Orbit type
import seaborn as sns
import matplotlib.pyplot as plt

sns.catplot(
    x="FlightNumber", 
    y="Orbit", 
    hue="Class", 
    data=df, 
    aspect=2, 
    kind="strip"
)

plt.xlabel("Flight Number", fontsize=14)
plt.ylabel("Orbit Type", fontsize=14)
plt.title("Flight Number vs Orbit Type by Success Class", fontsize=16)
plt.show()
```

You should see that in the LEO orbit the Success appears related to the number of flights; on the other hand, there seems to be no relationship between flight number when in GTO orbit.

R: En LEO, el éxito del aterrizaje mejora con la experiencia de lanzamiento (más vuelos).
   En GTO, no existe una relación evidente con el número de vuelos, indicando que otras variables técnicas pueden tener mayor influencia en el resultado.

### TASK  6: Visualize the launch success yearly trend


You can plot a line chart with x axis to be <code>Year</code> and y axis to be average success rate, to get the average launch success trend. 


The function will help you get the year from the date:


```python
# Extraer años desde la columna "Date"
def Extract_year(date_col):
    year = []
    for i in date_col:
        year.append(i.split("-")[0])  # Toma solo el año
    return year

df['Year'] = Extract_year(df["Date"])
```

```python
# Plot a line chart with x axis to be the extracted year and y axis to be the success rate
yearly_success = df.groupby('Year')['Class'].mean().reset_index()
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,6))
sns.lineplot(x='Year', y='Class', data=yearly_success, marker='o')

plt.title("Tendencia de éxito de lanzamiento por año", fontsize=16)
plt.xlabel("Año", fontsize=14)
plt.ylabel("Tasa de éxito promedio", fontsize=14)
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

You can observe that the success rate since 2013 kept increasing till 2017 (stable in 2014) and after 2015 it started increasing.

La tasa de éxito comienza a aumentar significativamente a partir de 2015.
Antes de eso, hubo lanzamientos con resultados más variables.
